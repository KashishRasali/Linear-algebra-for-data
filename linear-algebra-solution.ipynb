### 1. Matrix Operations

Given:  
A = [[2, -1], [0, 3]]  
B = [[1, 2], [-3, 4]]

#### (a) A + B
A + B = [[3, 1], [-3, 7]]

#### (b) AB
AB = [[5, 0], [-9, 12]]

#### (c) 3A − 2B
3A − 2B = [[4, -7], [6, 1]]
